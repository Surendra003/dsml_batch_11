1. Create a python program using at least 10 string from the table below and show their outputs with example.

```python
#.lower()
#.islower()
input = "There are more than 8 students in my 11th batch class"
for_lower= input.lower()
print(f"Lower of input is:\n{for_lower}")
is_lower=input.islower()
print(f"Checking Lower or not\n{is_lower}")
```

```python
#.upper()
#.isupper()
input = "There are more than 8 students in my 11th batch class"
for_upper = input.upper()
print(f"Upper of the input is:\n{for_upper}")
is_upper = input.isupper()
print(f"Checking Upper or not\n{is_upper}")
```

```python
#.capitalize() : capitalize the first letter
#.title() : capitalize the first letter of every word
input = "there are more than 8 students in my 11th batch class"
capitalization = input.capitalize()
print(f"Capitalization of first letter is:\n{capitalization}")
titaling = input.title()
print(f"Capatalizing every word:\n{titaling}")
```

```python
#.strip() : Removes whitespace from both ends
input = " There are more than 8 students in my 11th batch class "
no_whitespace = input.strip()
print(f"White space is removed:\n{no_whitespace}")
#.lstrip() : Removes left whitespace
no_left_whitespace = input.lstrip()
print(f"Left whitespace is removed:\n{no_left_whitespace}")
#.rstrip() : Removes right whitespace
no_right_whitespace = input.rstrip()
print(f"Right whitespace is removed:\n{no_right_whitespace}")

```

```python
#.replace(old,new)
input = "Hello Khan"
replacing = input.replace("Hello","Hi")
print(f"Replacing old 'Hello' with 'Hi':\n{replacing}")
#.split(seperator) : Split string into a list using seperator
spliting = input.split()
print(f"Spliting string into a list:\n{spliting}")
#.join(iterable) : joins element of a list eith a string iterable
tojoin = "Hello", "my", "name", "is Surendra."
joining = " ".join(tojoin)
print(f"Joining element of a list:\n{joining}")
```

```python
#.find() : provides the lowest index (position) where the substring is first found
#returns sstring of first occurance and return -1 if not found
tojoin = "Hello my name is Surendra."
found = tojoin.find("Surendra")
print(f"String found in: {found}")
not_found = tojoin.find("hello")
print(f"Not found: {not_found}")

#.index() : like function '.find()' but return error if not found
my = "Hello guys."
indexing = my.index("guys")
print(f"Index of string is: {indexing}")
```

```python
#.count() : counts how many times a substring occurs
my_introduction = "Hello this is my introduction. Also this is a platform for learning."
counting = my_introduction.count("this")
print(f"Count of the word is: {counting}")
```

```python
#.startswith(sub) : Checks if string start with a substring
my_introduction = "Hello this is my introduction. Also this is a platform for learning."
print(my_introduction)
substring = my_introduction.startswith("Also")
print(f"Is 'Also' a substring: {substring}")
substring1 = my_introduction.startswith("Hello")
print(f"Is 'Hello' a substring: {substring1}")
#.endingswith(sub) : Checks if string end with a substring
endingwith = my_introduction.endswith("learning")
print(f"Is learning a endstring: {endingwith}")
endingwith = my_introduction.endswith("learning.")
print(f"Is learning a endstring: {endingwith}")
```

```python
#.isalnum() : Returns True if all character are alphanumeric
a = "Hello my friend."
check1 = a.isalnum()
print(f"Is '{a}' alphanumeric: {check1}")
b = "Hello123"
check2 = b.isalnum()
print(f"Is '{b}' alphanumeric: {check2}")
b = "Hello12@"
check2 = b.isalnum()
print(f"Is '{b}' alphanumeric: {check2}")
#.isalpha() : Returns True if all character are alphabetic
a = "Hellomyfriend"
check1 = a.isalpha()
print(f"Is '{a}' alphabetic: {check1}")
a = "Hello my friend"
check1 = a.isalpha()
print(f"Is '{a}' alphabetic: {check1}")
#.isdigit() : Returns True if all character are digit
b = "12234"
check2 = b.isdigit()
print(f"Is '{b}' digit: {check2}")
#.isspace() : Returns True if all character are whitespace
check1 = " ".isspace()
print(check1)
#.islower() : Checks if all cased character are lower
a = "Hellomyfriend"
check1 = a.islower()
print(f"Is '{a}' lower: {check1}")
a = "hello my friend"
check1 = a.islower()
print(f"Is '{a}' lower: {check1}")
#.isupper() : Checks if all cased character are upper
a = "Hellomyfriend"
check1 = a.isupper()
print(f"Is '{a}' upper: {check1}")
a = "Hello My Friend"
check1 = a.isupper()
print(f"Is '{a}' upper: {check1}")
a = "HELLO"
check1 = a.isupper()
print(f"Is '{a}' upper: {check1}")
```

2. Take a sentence from the user and perform different string operation on it such as converting case, counting word,checking digit/alphabet, and replacing words. Print all results clearly with labels.

```python
#upper connversion
sentence = str(input("Enter a sentence you would like to express: "))
#checking if sentences are in upper or not
print(f"is '{sentence}' are in upper:",sentence.isupper())
#converting to upper
convert_to_upper = sentence.upper()
#displaying sentence after coversion to upper
print(f"Your sentence are converted to upper: {convert_to_upper}")
#checking if sentences are in upper or not after conversion
print(f"is'{convert_to_upper}' are in upper:",convert_to_upper.isupper())
#checking if sentences are in lower or not
print(f"Is '{sentence}' are in lower:",sentence.islower())
#conversing into lower
convert_to_lower = sentence.lower()
print(f"Your sentence are converted to lower: {convert_to_lower}")
```

```python
#capitalize conversion
text = str(input("Enter a sentence here: "))
capitalized = text.capitalize()
print(f"Converting '{text}' into first letter capital: {capitalized}")
#title conversion
titled = text.title()
print(f"Converting '{text}' into Title form: {titled}")
```
