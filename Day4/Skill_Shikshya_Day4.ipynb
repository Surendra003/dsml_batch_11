```python
#deep copy and shallow copy
#shallow copy
lst1=[[1,2,3,4,5],[37,38]]
lst2=lst1.copy()
lst1[1][1]=100
print(lst1)
print(lst2)
```

```python
#deep copy
import copy
lst1=[[1,2,3,4,5],[37,38]]
lst2=copy.deepcopy(lst1)
lst1[1][1]=100
print(lst1)
print(lst2)
#single ma deep copy and shallow copy is same
```
