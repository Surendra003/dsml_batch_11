```python
#different function in tuples
text = (1,3,4,5,6,"he",[3,4,"sa"])
print(text)
print(len(text))
num = (1,3,4,34,5,6)
print(len(num))
print(max(num))
print(min(num))
print(sum(num))
print(any(num))
print(all(num))
print(sorted(num))
```

```python
#unpacking
data =(1,2,3,4)
length,breadth,height,_ =data
print(f"a is:{length},b is {breadth},c is{height}")
```

```python
#indexing
t= (10,20,3040,3,4,5)
print(t[2])

#slicing
print(t[1:])

#concatinate
t1 =(40,60)
print(t+t1)

#Repetitions
t=(2,3)
print(t*3)

#membership in tuples
t=(10,20,30)
print(40 in t)

#comparision
t1=(1,2,3)
t2=(1,2,3)
print(t1==t2)

#wwe cannot remove individual element in tuple, instead delete whole tuple
tu = (1,2,3,4)
del tu
```

```python
#Set
sample_set = {"Surendra","Pritika",1,2}
print(sample_set)
print(sample_set)

set={}
print(type(set))
setA = {1,2,3,4}
setB = {2,3,6,7}
print(setA - setB)    #difference(-)
print(setA | setB)    #Union(|)
print(setA & setB)    #intersection(&)

print(setA^setB)      #symmetric difference(^)
print(4 in setA)      #Membership(in)
print(setA<=setB)
```
