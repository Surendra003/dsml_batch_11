```python
#task2: Entering age if the age is abouve 40 and have membership then gets 60% discount
#if the age is above 40 but doesnot have membership then gets 40% discount
age = int(input("Enter your age: "))
if age<=40:
  print("You have to pay full")
elif age>40:
  is_membership=input("Do you have membership: ")
  price=12000
  if is_membership.upper()=="YES":
    print(f"You got 60% discount: {price-(price*60)/100} out of {price}")
  elif is_membership.upper()=="NO":
    print(f"You got 40% discount: {price-(price*40)/100} out of {price}")
```

```python
#condition : there is list and if the user input is even then add user input
#if odd then remove 1 item from list
list = [3,4,5,6,7]
number = int(input("Enter a number:"))
if number%2==0:
  print("Even")
  list.append(number)
  print(list)
else:
  print("Odd")
  list.pop()
  print(list)
```
