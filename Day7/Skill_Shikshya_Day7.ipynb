```python
#to check students marks in grading system
marks = float(input("Enter your marks in grade: "))
if marks>3.6 and marks <=4.0:
  print("A+")
elif marks>3.2 and marks <=3.6:
  print("A")
elif marks>2.8 and marks <=3.2:
  print("B+")
elif marks>2.4 and marks <=2.8:
  print("A")
elif marks>2.0 and marks <=2.4:
  print("B+")
elif marks>1.6 and marks <=2.0:
  print("A")
elif marks<=1.6:
  print("NG")
else:
  print("Your number is out of range, must in between 1-4.")
```
